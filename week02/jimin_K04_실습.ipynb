```python
import numpy as np
import pandas as pd

data_df = pd.read_excel('/content/Boston_Housing.xlsx')
input_data = data_df[['RM', 'LSTAT']].to_numpy() #특성 두가지를 넘파이 배열로
target_df = data_df['MEDV'] # 주택 가격
```

```python
target_mean = target_df.mean()
target_data = [1 if target_df[i].item() > target_mean else 0 for i in range(0,506)] #주택가격이 평균이상이면1 아니면 0
print(target_data)
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(input_data, target_data, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled, train_target)
```

```python
print(lr.predict(test_scaled))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(test_target, lr.predict(test_scaled)))
print(lr.predict_proba(test_scaled[:10]))
```
